```python
buscaExecucaoVM = pd.read_excel(r'/content/drive/MyDrive/Projetos/wfm/Busca_Execucao/Vila/executados-vila.xlsx')
buscaExecucaoVM.head()
```

```python
buscaExecucaoVM.columns = ['OS', 'etapa', 'TSS', 'dsTSS', 'objOrigem',
       'iniciativa', 'solicitante', 'endereco', 'municipio',
       'TSE', 'dsTSE', 'observacao',
       'tempoExecucao', 'dataCompetencia', 'dataPlanejamento',
       'dataInicioExecucao', 'dataFimExecucao', 'inicioPausa',
       'fimPausa', 'tempoExecucao', 'fornecedor', 'contrato',
       'equipe', 'statusOperacao', 'statusOS']
```

```python
buscaExecucaoVM.head()
```

```python
# Criando um ID para deixar apenas a ultima iteracao da base (ordem-etapa):

buscaExecucaoVM['ordemTSS'] = buscaExecucaoVM['OS']+buscaExecucaoVM['TSS']
```

```python
# Alterando o tipo de dado da coluna para data:

buscaExecucaoVM['dataFimExecucao'] = pd.to_datetime(buscaExecucaoVM['dataFimExecucao'])
```

```python
# Deixando o dado mais recente no topo: 

buscaExecucaoVM = buscaExecucaoVM.sort_values('dataFimExecucao', ascending=True)
```

```python
buscaExecucaoVM = buscaExecucaoVM.drop_duplicates(subset='ordemTSS', keep='last')
```
